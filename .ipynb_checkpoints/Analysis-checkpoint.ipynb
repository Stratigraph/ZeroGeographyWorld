## Analysis functions

```python
def gini(list_of_values):
    '''
    Compute Gini coefficient for an array of values
    via:
        http://planspace.org/2013/06/21/how-to-calculate-gini-coefficient-from-raw-data-in-python/
    '''
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
    fair_area = height * len(list_of_values) / 2
    return (fair_area - area) / fair_area

def gini_series(interest_series):
    '''
    Compute the gini coefficient for each array in a time-series of arrays, and return
    the series of coefficients.
    '''
    gini_series = []
    for series in interest_series:
        gini_series.append(gini(series))
    return gini_series
```

```python
def alpha_series(interest_series):
    '''
    Fit a power law to each distribution in a series, and return series of alpha coefficients.
    '''
    alpha_series = []
    for series in interest_series:
        fit = powerfork.Fit(series, discrete=True)
        if np.isnan(fit.alpha):
            break
        alpha_series.append(fit.alpha)
    return alpha_series
```

# Analysis

For each experiment:

1. Visualize a sample model run, in terms of active agent counts
2. Show gini series across all model runs
3. Show alpha series across one and all runs

## Experiment 1

```python
data1 = BatchRunner.import_results("Outputs/Experiment1.out")
```

### Sample analysis

```python
i = 48
sample = data1[i]
```

```python
sample.keys()
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)

ax.plot(sample["active_agents"])
ax.set_xlabel("Step", fontsize=14)
ax.set_ylabel("# of Actors", fontsize=14)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot(111)
start_interests = sample['interests_per_turn'][0]
ax.bar(range(len(start_interests)), start_interests)
ax.set_xlabel("Actor", fontsize=14)
ax.set_ylabel("Interests Owned", fontsize=14)
ax.set_ylim(0, 45)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

end_interests = np.array(sample['interests_per_turn'][-1])
end_interests = end_interests[end_interests > 0]
ax.bar(range(len(end_interests)), end_interests)
ax.set_xlim(0,len(end_interests))


ax.set_xlabel("Actor", fontsize=14)
ax.set_ylabel("Interests Owned", fontsize=14)
ax.set_ylim(0, 45)
ax.grid(True)
```

```python

plt.plot(gini_series(sample["interests_per_turn"]))
```

```python
plt.plot(alpha_series(sample["interests_per_turn"]))
```

### All overlayed

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111)

for model in data1:
    s = gini_series(model["interests_per_turn"])
    ax.plot(s, color='b', alpha=0.01)

ax.set_xlabel("Step", fontsize=14)
ax.set_ylabel("Gini Coefficient", fontsize=14)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111)

for model in data1:
    s = alpha_series(model["interests_per_turn"])
    ax.plot(s, color='b', alpha=0.01)

ax.set_xlabel("Step", fontsize=14)
ax.set_ylabel("Alpha Coefficient", fontsize=14)
ax.grid(True)
```
