# Scenario Testing

Testing a few different scenarios: Base Model, Multipolar Model, and High-Resolution Model

# Base Model

100 interests, start with one interest per actor

```python
m = Model()
m.interest_count = 100
#m.internal_conflict_rule = "internal"
m.internal_conflict_rule = "external"
m.base_model()
for agent in m.agents:
    agent.grow_resources()
```

```python
agent_count = []
largest = []
second_largest = []
for i in range(1000):
    for j in range(10):
        m.interaction()
    # Count active agents
    count = len([len(agent.interests) for agent in m.agents
                       if len(agent.interests) > 0])
    agent_count.append(count)
    
    # Get size of second-largest agent
    sizes = [len(agent.interests) for agent in m.agents
                       if len(agent.interests) > 0]
    largest.append(max(sizes))
    sizes.remove(max(sizes))
    second_largest.append(max(sizes))
    
    for agent in m.agents:
        agent.grow_resources()
```

```python
plt.plot(agent_count)
```

```python
plt.plot(largest)
plt.plot(second_largest)
```

```python
interest_counts = np.array([len(agent.interests) for agent in m.agents])
interest_counts = interest_counts[interest_counts>0]
plt.bar(range(len(interest_counts)), interest_counts)
```

```python
plt.bar(range(len(m.agents)), [agent.wealth for agent in m.agents])
```

# Bipolar Model

Initialize the model into a bipolar setup, with two agents controlling 80% of interests between them.

```python
m = Model()
m.interest_count = 100
#m.internal_conflict_rule = "internal"
m.internal_conflict_rule = "external"
m.set_polarity(2, 0.8) # Two superpowers, with 80% of the world's interests between them.
#m.verbose = True
for agent in m.agents:
    agent.grow_resources()
```

```python
agent_count = []
largest = []
second_largest = []
for i in range(1000):
    for j in range(10):
        m.interaction()
    # Count active agents
    count = len([len(agent.interests) for agent in m.agents
                       if len(agent.interests) > 0])
    agent_count.append(count)
    
    # Get size of second-largest agent
    sizes = [len(agent.interests) for agent in m.agents]
    #largest.append(max(sizes))
    #sizes.remove(max(sizes))
    #second_largest.append(max(sizes))
    largest.append(sizes[0])
    second_largest.append(sizes[1])
    
    for agent in m.agents:
        agent.grow_resources()
```

```python
plt.plot(agent_count)
```

```python
plt.plot(largest)
plt.plot(second_largest)
```

```python
interest_counts = np.array([len(agent.interests) for agent in m.agents])
#interest_counts = interest_counts[interest_counts>0]
plt.bar(range(len(interest_counts)), interest_counts)
```

```python
plt.hist(interest_counts[interest_counts>0])
```
