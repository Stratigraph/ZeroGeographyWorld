## Analysis functions

```python
def gini(list_of_values):
    '''
    Compute Gini coefficient for an array of values
    via:
        http://planspace.org/2013/06/21/how-to-calculate-gini-coefficient-from-raw-data-in-python/
    '''
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
    fair_area = height * len(list_of_values) / 2
    return (fair_area - area) / fair_area

def gini_series(interest_series):
    '''
    Compute the gini coefficient for each array in a time-series of arrays, and return
    the series of coefficients.
    '''
    gini_series = []
    for series in interest_series:
        gini_series.append(gini(series))
    return gini_series
```

```python
def alpha_series(interest_series):
    '''
    Fit a power law to each distribution in a series, and return series of alpha coefficients.
    '''
    alpha_series = []
    for series in interest_series:
        fit = powerfork.Fit(series, discrete=True)
        if np.isnan(fit.alpha):
            break
        alpha_series.append(fit.alpha)
    return alpha_series
```
