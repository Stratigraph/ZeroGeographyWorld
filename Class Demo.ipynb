```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact
from Model import *
```

```python
%matplotlib inline
```

# Prep the model

```python
# Initialize the model
m = Model()
#m.internal_conflict_rule = 'external'
m.internal_conflict_rule = 'internal'
m.setup_model()
#m.verbose = True
for agent in m.agents:
    agent.grow_resources()
```

```python
# Prep data collection
agent_count = []
agent_wealths = []
agent_interests = []
```

```python
# Defining model run and draw function
def model_run(conflict_rule):
    global agent_count
    global agent_wealths
    global agent_interests
    global m
    
    agent_count = []
    agent_wealths = []
    agent_interests = []
    
    m = Model()
    m.internal_conflict_rule = conflict_rule
    m.setup_model()
    for agent in m.agents:
        agent.grow_resources()
    
    for i in range(500):
        if i == 0: continue
        # Run 5 iterations
        for j in range(5):
            m.interaction()
        # Count active agents
        interests = [len(agent.interests) for agent in m.agents
                           if len(agent.interests) > 0]
        agent_count.append(len(interests))
        agent_interests.append(interests)

        # Get the wealth of each agent
        wealths = [agent.wealth for agent in m.agents
                   if agent.interests > 0]
        agent_wealths.append(wealths)

        # Grow all agent resources
        for agent in m.agents:
            agent.grow_resources()
    
    # Visualize results
    fig1 = plt.figure(figsize=(12,6))
    ax = fig1.add_subplot(111)
    ax.plot(agent_count)
    ax.grid(True)
```

```python
def draw_histograms(tick):
    fig = plt.figure(figsize=(12,6))
    ax1 = fig.add_subplot(121)
    ax1.hist(agent_wealths[tick])
    ax1.set_title("Wealth distribution")
    ax1.set_xlabel("Wealth")
    ax1.set_ylabel("Count")
    ax1.grid(True)
    
    ax2 = fig.add_subplot(122)
    ax2.hist(agent_interests[tick])
    ax2.set_title("Interests Owned Distribution")
    ax2.set_xlabel("Interests Owned")
    ax2.set_ylabel("Count")
    ax2.grid(True)
```

# Visualize results

```python
interact(model_run,
         conflict_rule = ["internal", "external"])
```

```python
interact(draw_histograms,
         tick=(0,len(agent_count)-1))
```
